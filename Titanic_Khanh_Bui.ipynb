```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
path_to_file = '../../titanic.csv'
data = pd.read_csv(path_to_file, encoding='ISO-8859-1')
data
```

# 1a. What is the distribution of passenger ages?

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.drop("Cabin", axis=1, inplace=True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
for column in data: 
    if data[column].dtype != 'object':
        data[column] = data[column].fillna(data[column].mean())
    else:
        data[column] = data[column].fillna(data[column].mode()[0])
        
data
```

```python
data.isnull().sum()
```

# 1b. What is the passenger survival count based on gender?

```python
data
```

```python
survivors_by_sex = data[data['Survived'] == 1].groupby("Sex").count()
survivors_by_sex['Name']
```

# 2a. What is the count of passengers in each passenger class (Pclass)?

```python
data
```

```python
data.groupby(['Pclass'])['Name'].count()
```

# 2b. What is the survival rate for each gender (Sex)?

```python
survival_rate_by_gender = data.groupby('Sex')['Survived'].mean()
survival_rate_by_gender
```

# 3a. What is the survival rate based on both gender (Sex) and passenger class (Pclass)?

```python
survival_rate_by_class = data.groupby('Pclass')['Survived'].mean()
survival_rate_by_class
```

# 3b. What is the average fare paid by passengers based on their gender (Sex) and port of embarkation (Embarked)?

```python
data
```

```python
avg_fare_by_gender = data.groupby(['Sex']).Fare.mean()
avg_fare_by_gender
```

```python
avg_fare_by_embarked = data.groupby(['Embarked']).Fare.mean()
avg_fare_by_embarked
```
